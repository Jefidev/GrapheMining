# Arena jazz club

The arena jazz club materialize the collabortation between different jazz artist. We are going to analyze it

```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx

def graphFromEdgesFile (fname):
    G = nx.read_edgelist(fname, delimiter = '\t')
    return G

def drawGraph (G, pos = None):
    plt.figure (figsize=(15,15))
    if not pos:
        pos = nx.spring_layout(G)
    nx.draw_networkx(G, pos)
    plt.show()
    
    
Gj = graphFromEdgesFile('arenas-jazz/out.arenas-jazz')
drawGraph(Gj)
```
